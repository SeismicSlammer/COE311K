```python
import numpy as np
import matplotlib.pyplot as plt

# Data points 

X = np.array([1, 2, 3, 4, 5])

Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
plt.scatter(X, Y)
plt.plot(X, m*X + c)
```

```python
##PART 2

X_rand = np.random.rand(5)
Y_rand = np.random.rand(5)

A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T
m_rand, c_rand = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0]

print(f"Slope: {m_rand}, Intercept: {c_rand}")
plt.scatter(X_rand, Y_rand)
plt.plot(X_rand, m_rand*X_rand + c_rand)
```

```python
##With random data, the fit can be close or very much not close
```

```python
import numpy as np
import matplotlib.pyplot as plt

##equation (6x-7)
m_true = 6
c_true = -7

##using 20 points
x = np.linspace(0, 1, 20)

## pertubations
noise = np.random.normal(0, 0.5, len(x))
y = m_true * x + c_true + noise


A = np.vstack([x, np.ones(len(x))]).T
m_est, c_est = np.linalg.lstsq(A, y, rcond=None)[0]

print("True slope:", m_true)
print("Estimated slope:", m_est)
print("True intercept:", c_true)
print("Estimated intercept:", c_est)


plt.figure()
plt.scatter(x, y)
plt.plot(x, m_est * x + c_est)
plt.title("Least Squares Fit (20 Points)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
##The estimated slope and intercept are rather close, and I noticed that when playing around with more points,
##having more points increased the likelihood of a close estimate
```
