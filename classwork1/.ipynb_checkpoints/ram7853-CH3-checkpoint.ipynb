```python
import numpy as np

import matplotlib.pyplot as plt

def monteCarlo(sample):
    trials = 0
    inRadius = 0
    while trials < sample:
        x = np.random.rand()
        y = np.random.rand()
        radius = (x**2 + y**2)**.5
        if radius <= 1:
            inRadius += 1
        trials += 1
    piApprox = 4 *(inRadius/trials)
    return piApprox

absoluteError = []
relativeError = []
trialNumber = []

i=1

while i < 101:
    approximation = monteCarlo(1000)
    absoluteError.append(abs(np.pi-approximation))
    relativeError.append(abs((np.pi-approximation)/np.pi))
    trialNumber.append(i)
    
    i += 1
    
plt.subplot(2,1,1)
plt.plot(trialNumber, absoluteError, '.')
plt.xlabel("Trial Number")
plt.ylabel("Absolute Error")
plt.title("Errors vs Trial Number (1000 samples per trial)")
plt.subplot(2,1,2)
plt.plot(trialNumber, relativeError, '.')
plt.xlabel("Trial Number")
plt.ylabel("Relative Error")
plt.show()

    
```
