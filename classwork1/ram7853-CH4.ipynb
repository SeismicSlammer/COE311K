```python
import numpy as np

def leibniz(termNum):
    n = 0
    term = 0
    summation = 0
    while n < termNum:
        term = ((-1)**n) / (2*n + 1)
        summation += term
        n += 1
    return summation*4

print("Term ", 1, ": Error --", np.pi - leibniz(1))

for i in np.arange(0,5001,100):
    if i != 0:
        print("Term ", i, ": Error -- ", np.pi - leibniz(i))
        
        
        
        
        
```
