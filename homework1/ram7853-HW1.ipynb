```python
import numpy as np

import matplotlib.pyplot as plt

sampleSize = np.arange(10,1001,20)

def monteCarlo(sample):
    trials = 0
    inRadius = 0
    while trials < sample:
        x = np.random.rand()
        y = np.random.rand()
        radius = (x**2 + y**2)**.5
        if radius <= 1:
            inRadius += 1
        trials += 1
    piApprox = 4 *(inRadius/trials)
    return piApprox

errors = []

for i in sampleSize:
    approximation = monteCarlo(i)
    error = abs(np.pi-approximation)
    errors.append(error)
    
plt.scatter(sampleSize, errors)
plt.show

    
```
