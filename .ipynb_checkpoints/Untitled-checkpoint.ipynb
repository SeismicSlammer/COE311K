```python
import numpy as np

import matplotlib.pyplot as plt

def monteCarlo(sample):
    trials = 0
    inRadius = 0
    while trials < sample:
        x = np.random.rand()
        y = np.random.rand()
        radius = (x**2 + y**2)**.5
        if radius <= 1:
            inRadius += 1
        trials += 1
    piApprox = 4 *(inRadius/trials)
    return piApprox

absoluteError = []
relativeError = []

i=0

while i < 100:
    approximation = monteCarlo(1000)
    absoluteError.append(abs(np.pi-approximation))
    relativeError.append(abs((np.pi-approximation)/np.pi))
    
    i += 1
    
plt.scatter(absoluteError, relativeError)
plt.xlabel("Absolute Error")
plt.ylabel("Relative Error")
plt.show

    
```
