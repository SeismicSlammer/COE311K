```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10.1, .1)

def Sin_Taylor(terms):
    total = np.zeros_like(x)
    for n in range(terms):
        total += ((-1)**n) * x**(2*n + 1) / np.math.factorial(2*n + 1)
    
    return total

def absolute_Error(terms):
    approximation = Sin_Taylor(terms)
    errors = np.abs(np.sin(x) - approximation)
    return errors

plt.plot(x, absolute_Error(1), label="1 Term error")
plt.plot(x, absolute_Error(2), label="2 Term error")
plt.plot(x, absolute_Error(3), label="3 Term error")
plt.plot(x, absolute_Error(4), label="4 Term error")
plt.ylim(0,40)
plt.xlabel("x")
plt.ylabel("Absolute Error")

plt.legend()
plt.show()
```

```python
##Close to 0, the higher term functions have significantly less error, but the farther you get, the quicker the 
##higher term functions accumulate error, eventually overtaking the error of the lower term functions.
```
